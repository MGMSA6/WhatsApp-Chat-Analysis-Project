### Data preprocessing

```python
print(type(data))
```

```python
# Apply regular expression
pattern = '[[0-9]{2}\/[0-9]{2}\/[0-9]{2},\s[0-9]{1,2}:[0-9]{2}:[0-9]{2}\s[AaPp][Mm]]\s'
```

```python
message = re.split(pattern, data)
```

```python
# Create dataframe
df = pd.DataFrame({'user_message':message, 'message_date': dates})
```

```python
df.head(3)
```

```python
df['message_date'] = df['message_date'].str.replace(']','', regex=True)
```

```python
df['user_message'] = df['user_message'].str.replace('[','', regex=True)
```

```python
df.head()
```

```python
# Convert message_date to datetime type
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M:%S %p ')

df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df.head()
```

```python
type(df['date'])
```

```python
df.shape
```

```python
# Seperate users and message from dataframe

users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
def fetch_stats(selected_user, df):
    if selected_user != 'Overall':
        df = df[df['user'] == selected_user]

    # fetch the number fo messages
    num_messages = df.shape[0]

    # fetch the total numbers of words
    words = []
    for message in df['message']:
        words.extend(message.split())

    # fetch number of media messages
    num_media_messages = df[df['message'] == '\u200eimage omitted\n'].shape[0]

    # fetch number of links shared
    return num_messages, len(words), num_media_messages
```

```python
print(fetch_stats('Anand', df))
```

```python
period = []

for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str("00"))
    elif hour == 0:
        period.append(str("00") + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```
